```python
plt.figure(figsize=(9,7))
plt.plot(df.Year,df.USA,'b.-',label='united state')
plt.plot(df.Year,df.Australia,'r.-',label='Australia')
plt.plot(df['Year'],df['UK'],'g.-',label='UK')
plt.plot(df.Year,df.Canada,'y.-',label='Canada')
plt.title("Gas prices over time in (usd)",fontdict={'fontweight':'bold','fontsize':'22'})
#print(df.Year[::3])
plt.xticks(df.Year[::2].tolist()+[2011])
plt.xlabel('<-- --------------------------- Year -------------------- -->')
plt.ylabel('<-- ------------------------ Us dorlars  --------------------->')
plt.legend()
plt.grid()
plt.savefig('Gas_price_fig',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(9,7))
for country in df:
    if country!='Year':
            plt.plot(df.Year,df[country],marker='.')
plt.title("Gas prices over time in (usd)")
#print(df.Year[::3])
plt.xticks(df.Year[::2])
plt.xlabel('<-- --------------------------- Year -------------------- -->')
plt.ylabel('<-- ------------------------ Us dorlars  --------------------->')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,7))
country_look=['USA','UK','Canada','Australia']
for country in country_look:
    if country != 'Year':
        plt.plot(df.Year,df[country],marker='.',label=country_look)
plt.title("Gas prices over time in (usd)")
#print(df.Year[::3])
plt.xticks(df.Year[::2])
plt.xlabel('<-- --------------------------- Year -------------------- -->')
plt.ylabel('<-- ------------------------ Us dorlars  --------------------->')
plt.legend()
plt.grid()
plt.show()
```
